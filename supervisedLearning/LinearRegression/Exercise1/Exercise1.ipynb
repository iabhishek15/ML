```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('exercises-linear-regression-exercise-1/avg-household-size.csv')
df
```

```python
reg = linear_model.LinearRegression()
Y = df['avghouseholdsize']
X = df[['countyfips','statefips']]
reg.fit(X, Y)
```

```python
plt.scatter(df.countyfips, df.avghouseholdsize)
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = reg.coef_ * x_vals + reg.intercept_
plt.plot(x_vals, y_vals, '--', color = 'red')
```
